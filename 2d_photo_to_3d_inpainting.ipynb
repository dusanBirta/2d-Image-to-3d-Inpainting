#### Upload Image for generation

```python
from google.colab import files
import yaml
import os

# Navigate to '3d-photo-inpainting' directory
%cd /content/3d-photo-inpainting/image

# Delete existing image files in the '3d-photo-inpainting' directory
img_extensions = ['.png', '.jpg', '.jpeg', '.tiff', '.bmp', '.gif']
for img_file in os.listdir():
    if any(img_file.endswith(ext) for ext in img_extensions):
        os.remove(img_file)

# Upload new image
uploaded = files.upload()

# Extract the filename from the uploaded dictionary
for fn in uploaded.keys():
    # Save the uploaded file to '3d-photo-inpainting' directory
    with open(fn, 'wb') as f:
        f.write(uploaded[fn])
    print('User uploaded file "{name}" with length {length} bytes'.format(name=fn, length=len(uploaded[fn])))

# Navigate back to '/content/3d-photo-inpainting' directory
%cd /content/3d-photo-inpainting
```

#### Display Generated Videos

```python
# Get File name
uploaded_filename = list(uploaded.keys())[0]
# Remove File extension
uploaded_filename_without_extension = uploaded_filename.rsplit('.', 1)[0]
```

```python
from IPython.display import display, HTML
from base64 import b64encode

def play_video(video_path):
    # Load the video
    mp4 = open(video_path,'rb').read()
    data_url = "data:video/mp4;base64," + b64encode(mp4).decode()

    # Display the video
    return HTML("""
    <video width=400 controls>
          <source src="%s" type="video/mp4">
    </video>
    """ % data_url)

# Ending names of video files generated
video_names = ["_dolly-zoom-in.mp4", "_zoom-in.mp4", "_zoom-in.mp4", "_swing.mp4"]

for vid_name in video_names:
    video_path = f"/content/3d-photo-inpainting/video/{uploaded_filename_without_extension}{vid_name}"
    display(play_video(video_path))
```
